## Step 2 - Data Privacy

```python
# To drop first_name and Surname due to  personally identifiable information data concerns
df_ords_prods_custs = df_ords_prods_custs.drop(columns=['First Name', 'Surname'])
```

```python
# To print the first 5 rows after removing PII columns
df_ords_prods_custs.head()
```

## Step 3 - To compare customer behavior in different geographical areas

```python
# To Create a regional segmentation of the data by defining the for-loop statemrnt
region=[]
for value in df_ords_prods_custs['STATE']:
    if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
        region.append('Northeast')
    elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
        region.append('Midwest')
    elif value in ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana']:
        region.append('South')
    else:
        region.append('West')
```

```python
region
```

```python
# To combine result with df_orders_products_customers_merged
df_ords_prods_custs['region'] = region
```

```python
df_ords_prods_custs.head()
```

```python
# To know the frequency of “region” column 
df_ords_prods_custs['region'].value_counts(dropna = False)
```

```python
# To create cross tab for region and spend flag
df_crosstab_reg_spend = pd.crosstab(df_ords_prods_custs['region'], df_ords_prods_custs['spending_flag'], dropna = False)
```

```python
df_crosstab_reg_spend
```

```python
# Step 4 - To create an exclusion flag for low-activity customers
cust_activity = []
for value in df_ords_prods_custs["max_order"]:
      if value >= 5:
            cust_activity.append("High_activity customer")
      else:
        cust_activity.append("Low_activity customer")
```

```python
cust_activity
```

```python
# To combine cust_activity column with df_ords_prods_custs
df_ords_prods_custs['customer_activity'] = cust_activity
```

```python
df_ords_prods_custs.head()
```

```python
# To know the frequency of “customer_activity” column 
df_ords_prods_custs['customer_activity'].value_counts(dropna = False)
```

```python
# To create a subset for High-activity customers
df_high_cust_activity = df_ords_prods_custs[df_ords_prods_custs['customer_activity'] == 'High_activity customer']
```

```python
# To export high customer activity data
df_high_cust_activity.to_pickle(os.path.join(path,'02 Data','Prepared Data','High_activity_cust.pkl'))
```

# Step 5 - Profiling
## Profile: Age

```python
# Age frequency
df_ords_prods_custs['Age'].value_counts(dropna = False)
```

```python
# Step 4 - To create an Age group flag 
Age_group = []
for value in df_ords_prods_custs["Age"]:
    if value < 35:
        Age_group.append("Young_Adult")
    elif value >= 35 and value < 60:
        Age_group.append("Middle_aged")
    else:
        Age_group.append("Elderly")
```

```python
Age_group
```

```python
# To combine Age_group column with df_ords_prods_custs
df_ords_prods_custs['Age_group'] = Age_group
```

```python
df_ords_prods_custs['Age_group'].value_counts(dropna = False)
```

```python
# To create Bar chart for Age group
Bar_Age = df_ords_prods_custs['Age_group'].value_counts().plot.bar(color = ['green', 'purple', 'blue'])
```

```python
# To export Bar chart for Age group
Bar_Age.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Bar_Age.png'))
```

```python
# To aggregate the max, mean, and min values on a customer-profile level to show distribution
df_ords_prods_custs.groupby('Age_group').agg({'days_since_last_order':['max', 'mean', 'min'], 'prices':['max', 'mean', 'min']})
```

## Profile: income

```python
# Step 4 - To create a flag for income level
df_ords_prods_custs_dept.loc[df_ords_prods_custs_dept['income'] <= 50000, 'income_level'] = 'Low income'
df_ords_prods_custs_dept.loc[(df_ords_prods_custs_dept['income'] >50000) & (df_ords_prods_custs_dept['income'] <= 150000), 'income_level'] = 'Middle income'
df_ords_prods_custs_dept.loc[df_ords_prods_custs_dept['income'] > 150000, 'income_level'] = 'High income'
```

```python
df_ords_prods_custs_dept['income_level'].value_counts(dropna = False)
```

```python
# To create bar chart for income level
income_bar = df_ords_prods_custs_dept['income_level'].value_counts().plot.bar(color = ['orange', 'green', 'blue'])
```

```python
# To export income level bar chart

income_bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_chart_income_level'))
```

```python
# To aggregate the max, mean, and min values on a customer-profile income level to show distribution
income_agg = df_ords_prods_custs_dept.groupby('income_level').agg({'days_since_last_order': ['min', 'max', 'mean'], 'prices': ['min', 'max', 'mean', 'sum']})
```

```python
income_agg
```

customers with low earning power have the lowest mean price for orders, and a longer wait time before they place their next order. Middle income customers generate the majority of sales for instacart

### Profile: number of dependants

```python
# To create family_status_flag for each customer: 
df_ords_prods_custs_dept.loc[(df_ords_prods_custs_dept['Family_status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_ords_prods_custs_dept['n_dependants']==0), 'family_level'] = 'single no dependants'
df_ords_prods_custs_dept.loc[(df_ords_prods_custs_dept['Family_status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_ords_prods_custs_dept['n_dependants'] >=1), 'family_level'] = 'single with dependants'
df_ords_prods_custs_dept.loc[(df_ords_prods_custs_dept['Family_status'] == 'married') & (df_ords_prods_custs_dept['n_dependants'] == 1), 'family_level'] = 'married no dependants'
df_ords_prods_custs_dept.loc[(df_ords_prods_custs_dept['Family_status'] == 'married') & (df_ords_prods_custs_dept['n_dependants'] > 1), 'family_level'] = 'married with dependants'
```
