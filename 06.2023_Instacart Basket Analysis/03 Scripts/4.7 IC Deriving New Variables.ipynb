# To use user-based function for price range

```python
# Define the function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# To check the values in new column
df['price_range'].value_counts(dropna = False)
```

```python
# To check the maximum number in the price column
df['prices'].max()
```

# Using If-Statements with For-Loops

```python
df_orders_products_merged
```

```python
# To know the frequency of “orders_day_of_the_week” column 
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Define the for-loop statemrnt
result = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# To combine result with df_orders_products_merged
df_orders_products_merged['Busiest day'] = result
```

```python
# To know the frequency of “orders_day_of_the_week” column 
df_orders_products_merged['Busiest day'].value_counts(dropna = False)
```

# Task 4.7

```python
# Step 2 - To Define the for-loop statemrnt
results = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    results.append("Busiest days")
  elif value == 3 or value == 4:
    results.append("Least busy days")
  else:
    results.append("Regularly busy")
```

```python
results
```

```python
# To combine results with df_orders_products_merged
df_orders_products_merged['Busiest days'] = results
```

```python
# To know the frequency of “orders_day_of_the_week” column 
df_orders_products_merged['Busiest days'].value_counts(dropna = False)
```

```python
# To Define the for-loop statement for busiest_period_of_day
Hours = []

for value in df_orders_products_merged["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    Hours.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    Hours.append("Fewest orders")
  else:
    Hours.append("Average orders")
```
